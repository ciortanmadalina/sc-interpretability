```python
import sys
sys.path.append("..")
import argparse
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from sklearn.cluster import KMeans
from sklearn import metrics

import torch
import torch.nn as nn
import copy
from tqdm.notebook import tqdm
import models
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import st_loss
import time
import h5py
import scipy as sp
import scanpy.api as sc
from collections import Counter
import random
import utils
import interpret_utils
import seaborn as sns
import pickle

print(plt.style.available)
import matplotlib
matplotlib.style.use('seaborn')

import train
import os
import glob2
plt.ion()
plt.show()
%load_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
sc_dropouts = {
    '1c': 0.3,
    '-1c': 0.08,
    '0c': 0.17
}
```

```python
path = "../"
# category = "balanced_data"
category = "imbalanced_data"
files = glob2.glob(f'{path}R/simulated_data/{category}/*.h5')
files = [f[len(f"{path}R/simulated_data/{category}/"):-3] for f in files]
len(files), files
```

```python
methods = ["scziDesk","contrastivesc" ,"scDeepCluster",   "truth",]
```

```python
fm = {0: "3. ACC all DE genes", 30: "1. ACC top 30 DE genes", 100: "2. Acc top 100 DE genes"}

nb_features = -1
all_df = pd.DataFrame()
timeDf= pd.DataFrame()
for category in [ "imbalanced_data", "balanced_data", ]:
    for method in methods:
        print(f"{category}: {method}")
        df = pd.DataFrame()

        for k, nb_features_truth in enumerate([0, 30, 100]):
            for dataset in files:
                tokens = dataset.split("_")
                dropout= sc_dropouts[tokens[1]]
                de = tokens[-1]
                folder = f"../output/interpretability/{category}/{method}"
                with open(
                        f"../output/interpretability/{category}/{method}/{dataset}_all.pkl",
                        'rb') as f:
                    results = pickle.load(f)
                ari = round(results["meta"]["full_extra"]["ari"], 2)
                results = interpret_utils.addR(results, method, nb_features, category, dataset)

                inp1 = {k:v for k,v in results["features"].items() if k.startswith("full") == False}
                inp2 = {k:v for k,v in results["features"].items() if k.startswith("full")}
                inp2["truth"] = results["features_complete"]["truth_complete"]
                matrices1 = interpret_utils.analyze(inp1, None, nb_features_truth )
                matrices2 = interpret_utils.analyze(inp2, None, nb_features_truth )

                df1 = pd.concat([matrices1["acc"], matrices2["acc"]], axis = 1)

                df1["clusters"] = len(df1)
                df1["dataset"] = dataset
                df1["Dropout"] = dropout
                df1["de"] = float(de)
                df1["ARI"] = ari
                df1["method"] = method
                df1["category"] = category
                df1["nb_features_truth"] = nb_features_truth
                df1["nb_features_truth_label"] = fm[nb_features_truth]
                df = df.append(df1, ignore_index = True)
                
                results["time"] ={** results["time"], **{"dataset": dataset, "category": category, "method": method}}
                timeDf = timeDf.append(results["time"], ignore_index = True)
#                 break
#             break
#         break
#     break
            df.to_excel(f"../output/interpretability/{category}_{method}.xlsx")
            all_df = all_df.append(df, ignore_index = True)
```

# Figure 2

```python
full_idx = np.array([1,3,5,7,9,11, 13])
labels = ["(a) ACC\nbalanced\ndata\nall DE\ngenes", "(b) ACC \nimbalanced \ndata\nall DE\ngenes"]
methods = ["scziDesk", "scDeepCluster",  "contrastivesc" , "truth"]
plt.figure(figsize = (13, 13))
for cc, category in enumerate( [ "balanced_data" , "imbalanced_data",]):
    for k, method in enumerate(methods):
        group =all_df[(all_df["method"] == method) & (all_df["nb_features_truth"] == 0)
                     & (all_df["category"]==category)]

        ranks = pd.DataFrame(columns = new_cols, data = group[cols].values)
        factor = 1 if cc ==0 else 5
        ax = plt.subplot(3, 4, k+factor)
        plt.title(f"{method}")
        if k == 0:
            plt.text(-5, 0.485, labels[cc], fontweight = "bold")
        if method == "truth":
            ranks[ranks.columns[full_idx]].boxplot(ax = ax)
        else:
            ranks.boxplot(ax = ax)
            plt.gca().get_xticklabels()[14].set_fontweight('bold') 
            plt.gca().get_xticklabels()[15].set_fontweight('bold') 
        plt.xticks(rotation = 90)
        plt.ylabel("Accuracy")
        plt.ylim((0, 0.7))

for k, method in enumerate(['scziDesk', 'scDeepCluster', 'contrastivesc', "truth"]):
    group =all_df[(all_df["method"] == method)]

    ranks = pd.DataFrame(columns = new_cols, data = group[cols].values)
    ax = plt.subplot(3, 4, k+ 9)
    plt.title(f"{method}")
    ranks["nb_features_truth_label"] = group["nb_features_truth_label"].values
    if k == 0:
        plt.text(-5, 0.52, "(c) Mean\nglobal\nACC", fontweight = "bold")
    if method == "truth":
        ranks.groupby("nb_features_truth_label").mean()[new_cols][new_cols[full_idx]].T.plot.bar(ax = ax)
    else:
        ranks.groupby("nb_features_truth_label").mean()[new_cols].T.plot.bar(ax = ax)
        plt.gca().get_xticklabels()[14].set_fontweight('bold') 
        plt.gca().get_xticklabels()[15].set_fontweight('bold') 
    if k == 3:
        plt.legend(title= "Feature selection threshold")
    else:
        plt.legend("")

    plt.xticks(rotation = 90)
    plt.ylim((0, 0.65))
    plt.ylabel("Accuracy")

plt.tight_layout()
plt.savefig(f"../diagrams/global_boxplot.pdf", bbox_inches='tight')
```
